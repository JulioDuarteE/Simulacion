```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución exponencial
lambda_ = 1/15  # tasa (1 por cada 15 minutos)
horas_operacion = 12
minutos_totales = horas_operacion * 60

# Simulación de tiempos entre llegadas
tiempos_entre_llegadas = []
tiempo_acumulado = 0

while tiempo_acumulado < minutos_totales:
    tiempo = np.random.exponential(scale=1/lambda_)
    tiempos_entre_llegadas.append(tiempo)
    tiempo_acumulado += tiempo

# Eliminamos el último tiempo si excede las 12 horas
if tiempo_acumulado > minutos_totales:
    tiempos_entre_llegadas = tiempos_entre_llegadas[:-1]

# Convertimos a array de numpy para facilidad de cálculo
tiempos_entre_llegadas = np.array(tiempos_entre_llegadas)

# Resultados
print(f"Total de vehículos llegados en {horas_operacion} horas: {len(tiempos_entre_llegadas)}")
print(f"Tiempo promedio entre llegadas: {np.mean(tiempos_entre_llegadas):.2f} minutos")
print(f"Tiempo mínimo entre llegadas: {np.min(tiempos_entre_llegadas):.2f} minutos")
print(f"Tiempo máximo entre llegadas: {np.max(tiempos_entre_llegadas):.2f} minutos")

# Gráfico de los tiempos entre llegadas
plt.figure(figsize=(10, 5))
plt.plot(tiempos_entre_llegadas, 'o-')
plt.title('Tiempos entre llegadas de vehículos eléctricos')
plt.xlabel('Número de llegada')
plt.ylabel('Minutos desde llegada anterior')
plt.grid(True)
plt.show()

# Histograma de los tiempos entre llegadas
plt.figure(figsize=(10, 5))
plt.hist(tiempos_entre_llegadas, bins=30, edgecolor='black')
plt.title('Distribución de tiempos entre llegadas')
plt.xlabel('Minutos entre llegadas')
plt.ylabel('Frecuencia')
plt.axvline(x=15, color='r', linestyle='--', label='Media teórica (15 min)')
plt.legend()
plt.grid(True)
plt.show()
```

#Simulacion


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Paso 1: Generar 100 números aleatorios uniformemente distribuidos en [0, 1]
np.random.seed(42)  # Fijar la semilla para reproducibilidad
random_numbers = np.random.rand(100)

# Paso 2: Aplicar la prueba de Kolmogorov-Smirnov
# La hipótesis nula es que los números siguen una distribución uniforme en [0, 1]
D, p_value = stats.kstest(random_numbers, 'uniform')

# Mostrar los resultados
print(f"Estadístico de Kolmogorov-Smirnov: {D}")
print(f"Valor p: {p_value}")

# Interpretación del valor p
alpha = 0.05  # Nivel de significancia
if p_value > alpha:
    print("No se puede rechazar la hipótesis nula: los datos parecen seguir una distribución uniforme.")
else:
    print("Se rechaza la hipótesis nula: los datos no siguen una distribución uniforme.")

#  Graficar la distribución empírica y la distribución teórica
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=10, density=True, alpha=0.6, color='g', label='Distribución empírica')
plt.plot([0, 1], [1, 1], 'r-', label='Distribución teórica uniforme')
plt.title('Prueba de Kolmogorov-Smirnov para uniformidad')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```
